```python
from IPython.core.display import HTML, display
from captum.attr import visualization as viz
import json
```

```python
def print_aggregated(target,aggregated):    
    to_print=""
    to_print = to_print+"<html><body>"
    x=viz.format_word_importances([t for t,a in aggregated],[a for t,a in aggregated])
    to_print = to_print+"<b>"+str(target)+"</b>"
    to_print = to_print+"""<table style="border:solid;">"""+x+"</table>"
    to_print = to_print+"</body></html>"
    display(HTML(to_print))
```

```python
def visualize_essay(essay):
    if len(essay)==3:
        gold = essay[0]
        pred = essay[1]
        aggs = essay[2] # list of aggregates for grades 1-5
        print("Prediction", pred)
    elif len(essay)==2:
        gold = essay[0]
        aggs = essay[1]
    else:
        raise ValueError
    print("Gold:", gold+1)
    for i, agg in enumerate(aggs):
        print_aggregated(str(i+1), aggs[i])
```

### Layer: bert.embeddings

```python
with open("aggregates.json", "rt") as f:
    aggregates = json.load(f)
    
print("Number of essays:", len(aggregates))
# This one is grade 5

visualize_essay(aggregates[5])
```
